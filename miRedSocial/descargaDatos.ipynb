```python
import os
import django

os.environ.setdefault("DJANGO_SETTINGS_MODULE", "miRedSocial.settings")
django.setup()
print("Django inicializado correctamente")
```

```python
from asgiref.sync import sync_to_async
from django.contrib.auth.models import User
import pandas as pd

usuarios = await sync_to_async(list)(User.objects.all())

data = []

for usr in usuarios:
    data.append({
        "id":usr.id,
        "username":usr.username,
        "email":usr.email,
        "first_name":usr.first_name,
        "last_name":usr.last_name,
        "password_hash":usr.password,
        "is_active":True,
        "is_staff":True,
        "is_superuser": True
    })

df = pd.DataFrame(data)
df
```

```python
df.to_csv("UsuariosExportados.csv", index=False)
```
